### Mean Shift Clustering Algorithm

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Data Collection

```python
dataset = pd.read_csv("Mall_Customers.csv")
dataset
```

### Input Selection

```python
X = dataset.iloc[:,[3,4]].values
X
```

### Model Creation

```python
from sklearn.cluster import MeanShift

# Parameters 
            # bandwidth=None,  Bandwidth used in the flat kernel.
            # seeds=None, bin_seeding=False, min_bin_freq=1, cluster_all=True, n_jobs=None, max_iter=300)

MShift_Model = MeanShift(bandwidth=5)
MShift_Model.fit(X)
```

```python
dir(MShift_Model)
```

### Predictions

```python
Prediction= MShift_Model.labels_
Prediction
```

```python
dataset
```

### Predicted Clusters added to the new table

```python
Supervised = dataset
Supervised['Cluster_Group'] = Prediction
Supervised
```

### Visualize the model

```python
ClustersView = sns.scatterplot(data=Supervised, x=Supervised.columns[3], y=Supervised.columns[4], hue=Supervised.columns[5], palette='bright') 
# palette = bright, deep, muted, dark, pastel, colorblind
ClustersView
```
