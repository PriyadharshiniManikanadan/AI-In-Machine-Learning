### Hierarchical Clustering Algorithm

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Data Collection

```python
dataset = pd.read_csv("Mall_Customers.csv")
dataset
```

### Input Selection

```python
X = dataset.iloc[:,[3,4]].values
X
```

### Dendrogram

```python
import scipy.cluster.hierarchy as sch
Dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel("Clusters")
plt.ylabel("Euclidean Distance")
plt.show()    
```

### Model Creation

```python
from sklearn.cluster import AgglomerativeClustering
Clus_model = AgglomerativeClustering(n_clusters=5)
Prediction = Clus_model.fit_predict(X)
```

### Predicted Clusters Group

```python
Prediction
```

```python
dataset
```

### Predicted Clusters added to the new table

```python
Supervised = dataset
Supervised['Cluster_Group'] = Prediction
Supervised
```

### From Table to CSV file

```python
Supervised.to_csv("Dendrogram_Clusters.csv",index = False)
```
