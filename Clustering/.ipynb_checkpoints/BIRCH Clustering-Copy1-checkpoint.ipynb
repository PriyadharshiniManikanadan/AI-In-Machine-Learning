### OPTICS Clustering Algorithm (Ordering Points To Identify Cluster Structure)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Data Collection

```python
dataset = pd.read_csv("Mall_Customers.csv")
dataset
```

### Input Selection

```python
X = dataset.iloc[:,[3,4]].values
X
```

### Model Creation

```python
from sklearn.cluster import OPTICS

# Parameters 
            #  min_samples=5, max_eps=inf, metric='minkowski', p=2, metric_params=None, cluster_method='xi', eps=None,predecessor_correction=True, min_cluster_size=None, leaf_size=30, memory=None, n_jobs=None)
            # algorithm= {â€˜autoâ€™, â€˜ball_treeâ€™, â€˜kd_treeâ€™, â€˜bruteâ€™}
            # xi=0.05 - Determines the minimum steepness on the reachability plot that constitutes a cluster boundary.

OPTICS_Model = OPTICS(min_samples=2, xi=0.05, min_cluster_size=0.1)
OPTICS_Model.fit(X)
```

```python
dir(OPTICS_Model)
```

### Predictions

```python
Prediction= OPTICS_Model.labels_
Prediction
```

```python
dataset
```

### Predicted Clusters added to the new table

```python
Supervised = dataset
Supervised['Cluster_Group'] = Prediction
Supervised
```

### Visualize the model

```python
ClustersView = sns.scatterplot(data=Supervised, x=Supervised.columns[3], y=Supervised.columns[4], hue=Supervised.columns[5], palette='bright') 
# palette = bright, deep, muted, dark, pastel, colorblind
ClustersView
```
