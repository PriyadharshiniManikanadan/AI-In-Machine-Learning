### Affinity Propagation Clustering Algorithm

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Data Collection

```python
dataset = pd.read_csv("Mall_Customers.csv")
dataset
```

### Input Selection

```python
X = dataset.iloc[:,[3,4]].values
X
```

### Model Creation

```python
from sklearn.cluster import AffinityPropagation

# Parameters 
            # damping : float, default=0.5
            # max_iter : int, default=200
            # affinity : {‘euclidean’, ‘precomputed’}, default=’euclidean’

Aff_Model = AffinityPropagation(random_state=42)
Aff_Model.fit(X)
```

```python
dir(Aff_Model)
```

### Predictions

```python
Prediction = Aff_Model.predict(X) 
# Aff_Model.labels_ --- both gives the same results
```

```python
Aff_Model.cluster_centers_
```

```python
dataset
```

### Predicted Clusters added to the new table

```python
Supervised = dataset
Supervised['Cluster_Group'] = Prediction
Supervised
```

### Visualize the model

```python
ClustersView = sns.scatterplot(data=Supervised, x=Supervised.columns[3], y=Supervised.columns[4], hue=Supervised.columns[5], palette='bright') 
# palette = bright, deep, muted, dark, pastel, colorblind
ClustersView
```
