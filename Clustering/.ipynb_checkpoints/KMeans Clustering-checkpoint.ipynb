### K-Means Algorithm

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Data Collection

```python
dataset = pd.read_csv("Mall_Customers.csv")
dataset
```

### Input Selection

```python
X = dataset.iloc[:,[3,4]].values
X
```

### Find the No oF Clusters

```python
from sklearn.cluster import KMeans
list1 = []
for i in range (1,11):
    kmeans = KMeans(n_clusters=i, init= 'k-means++', random_state= 42)  
    kmeans.fit(X)                                                       
    list1.append(kmeans.inertia_)
plt.plot(range(1,11),list1)
plt.title("Elbow Method")
plt.xlabel("No of Clusters")
plt.ylabel("WCSS") # Within Clusters Sum of Squares
plt.show()

# Set the number of clusters to i
# Use the k-means++ method for centroid initialization, (which helps to speed up convergence).
# By setting random_state=42, or any fixed integer, you ensure that every time you run the code,
# the randomization produces the same results
    
```

```python
list1
```

### Model Creation

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
```

### Predicted Clusters Group

```python
y_kmeans
```

```python
dataset
```

### Predicted Clusters added to the new table

```python
supervised = dataset
supervised['Cluster_Group'] = y_kmeans
supervised
```

### From Table to CSV file

```python
supervised.to_csv("Clusters.csv",index = False)
```
