### DBSCAN Clustering Algorithm ( Density-Based Spatial Clustering of Applications with Noise)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Data Collection

```python
dataset = pd.read_csv("Mall_Customers.csv")
dataset
```

### Input Selection

```python
X = dataset.iloc[:,[3,4]].values
X
```

### Model Creation

```python
from sklearn.cluster import DBSCAN

# Parameters 
            # eps=0.5, float, default=0.5         
                        # The maximum distance between two samples for one to be considered as in the neighborhood of the other. 
                        # This is not a maximum bound on the distances of points within a cluster. 
                        # This is the most important DBSCAN parameter to choose appropriately for your data set and distance function. 
           # algorithm {‘auto’, ‘ball_tree’, ‘kd_tree’, ‘brute’}, default=’auto’
           # min_samples=5, metric='euclidean', metric_params=None, algorithm='auto', leaf_size=30, p=None, n_jobs=None)

DBSCAN_Model = DBSCAN(eps=3, min_samples=6)
DBSCAN_Model.fit(X)
```

```python
dir(DBSCAN_Model)
```

### Predictions

```python
Prediction= DBSCAN_Model.labels_
Prediction
```

```python
dataset
```

### Predicted Clusters added to the new table

```python
Supervised = dataset
Supervised['Cluster_Group'] = Prediction
Supervised
```

### Visualize the model

```python
ClustersView = sns.scatterplot(data=Supervised, x=Supervised.columns[3], y=Supervised.columns[4], hue=Supervised.columns[5], palette='bright') 
# palette = bright, deep, muted, dark, pastel, colorblind
ClustersView
```
