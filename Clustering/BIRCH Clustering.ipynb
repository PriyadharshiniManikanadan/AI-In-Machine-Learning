### BIRCH Clustering Algorithm (Balanced Iterative Reducing & Clustering using Hierarchies)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Data Collection

```python
dataset = pd.read_csv("Mall_Customers.csv")
dataset
```

### Input Selection

```python
X = dataset.iloc[:,[3,4]].values
X
```

### Model Creation

```python
from sklearn.cluster import Birch

# Parameters 
            # threshold=0.5, branching_factor=50, n_clusters=3, compute_labels=True, copy=True

Birch_Model = Birch()
Birch_Model.fit(X)
```

```python
dir(Birch_Model)
```

### Predictions

```python
Prediction= Birch_Model.labels_
Prediction
```

```python
dataset
```

### Predicted Clusters added to the new table

```python
Supervised = dataset
Supervised['Cluster_Group'] = Prediction
Supervised
```

### Visualize the model

```python
ClustersView = sns.scatterplot(data=Supervised, x=Supervised.columns[3], y=Supervised.columns[4], hue=Supervised.columns[5], palette='bright') 
# palette = bright, deep, muted, dark, pastel, colorblind
ClustersView
```
