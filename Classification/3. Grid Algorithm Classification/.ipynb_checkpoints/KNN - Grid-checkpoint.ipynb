## Classification Algorithm
## K Nearest Neighbors - Grid

```python
import pandas as pd
```

### Data collection

```python
dataset = pd.read_csv("Social_Network_Ads.csv")
```

```python
dataset
```

### Data preprocessing

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset= dataset.drop('User ID',axis=1)
```

```python
dataset
```

### Check the No of Unique classes and its count

```python
dataset['Purchased'].value_counts()
```

### Input Output Split

```python
indep = dataset[['Age','EstimatedSalary','Gender_Male']]
dep = dataset[['Purchased']]
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(indep,dep,test_size=0.20,random_state=0)
```

### Standardization

```python
from sklearn.preprocessing import StandardScaler
SC = StandardScaler()
X_train = SC.fit_transform(X_train)
X_test = SC.transform(X_test)
```

### Train Set

### Model Creation in GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

param_grid = {'n_neighbors' : [5,10,100,500,1000,2000],
              'weights': ['uniform','distance'],
              'algorithm':['auto','ball-tree','kd_tree','bryte'],
              'metric': ['minkowski']}

KNN = GridSearchCV(KNeighborsClassifier(), param_grid, refit=True, verbose = 3, n_jobs = -1, scoring = 'f1_weighted')
KNN.fit(X_train,Y_train)
```

### Test Set

```python
results = KNN.cv_results_

grid_predictions = KNN.predict(X_test)
```

### Evaluation Metrics 
### Confusion Matix

```python
from sklearn.metrics import confusion_matrix
CM = confusion_matrix(Y_test,grid_predictions)
CM
```

### Classification Report

```python
from sklearn.metrics import classification_report
Clf_report = classification_report(Y_test,grid_predictions)
print(Clf_report)
```

### ADDITIONAL : Check for the best params & f1_score 

```python
from sklearn.metrics import f1_score
f1_macro = f1_score(Y_test,grid_predictions,average='weighted')
f1_macro
```

### ADDITIONAL : Check roc_auc_score -Receiver Operating Characterstic (Area Under Curve)

```python
# ROC_AUC : Is a performance metric used primarily for binary classification problems to evaluate the quality of a classification model.
# predict_proba : Return probability estimates for the test 

from sklearn.metrics import roc_auc_score
ROC_score = roc_auc_score(Y_test,KNN.predict_proba(X_test)[:,1])
ROC_score
```
