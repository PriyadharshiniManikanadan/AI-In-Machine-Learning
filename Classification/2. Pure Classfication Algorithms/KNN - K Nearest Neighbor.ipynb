# Machine Learning - Classification
## KNN - K Nearest Neighbor

```python
import pandas as pd
```

### Data Collection

```python
dataset = pd.read_csv("Social_Network_Ads.csv")
```

```python
dataset
```

### Data preprocessing

```python
dataset= pd.get_dummies(dataset,dtype = int,drop_first=True)
```

```python
dataset
```

```python
dataset.shape
```

```python
dataset = dataset.drop('User ID', axis=1)
```

### Check the No of Unique classes and its count

```python
dataset['Purchased'].value_counts()
```

```python
dataset
```

### Input Output split

```python
indep = dataset[['Age','EstimatedSalary','Gender_Male']]
dep = dataset[['Purchased']]
```

### Train Test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(indep,dep,test_size=0.20,random_state=0)
```

### Standardization

```python
from sklearn.preprocessing import StandardScaler
SC = StandardScaler()
X_train = SC.fit_transform(X_train)
X_test = SC.transform(X_test)
```

### Train Set

### Model Creation

```python
from sklearn.neighbors import KNeighborsClassifier
# Parameters
# n_neighborsint, default=5
# weights{‘uniform’, ‘distance’},                        Weight function used in prediction.
# algorithm{‘auto’, ‘ball_tree’, ‘kd_tree’, ‘brute’}     Algorithm used to compute the nearest neighbors:
# metric -  default=’minkowski’                          Default is “minkowski”, which results in the standard Euclidean distance when p = 2
Classifier = KNeighborsClassifier(n_neighbors=10,weights='distance')
Classifier.fit(X_train,Y_train)
```

### Test Set

```python
Y_pred = Classifier.predict(X_test)
```

### Evaluation Metrics 
### Confusion Matix

```python
from sklearn.metrics import confusion_matrix
CM = confusion_matrix(Y_test,Y_pred)
CM
```

### Evaluate -  Accuracy, Recall, Precision, F1 Score, Macro Average, Weighted Average

```python
from sklearn.metrics import classification_report
Clf_report = classification_report(Y_test,Y_pred)
print(Clf_report)
```
