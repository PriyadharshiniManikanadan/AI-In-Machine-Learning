```python
import pandas as pd 
```

### Data Collection

```python
dataset = pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

### Data Prepreocessing

```python
dataset= pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

### Input Ouput Split

```python
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes' ]]
dependent = dataset[['charges']]
```

```python
independent
```

```python
dependent
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(independent,dependent,test_size=0.2,random_state=0)
```

### Standardization

```python
from sklearn.preprocessing import StandardScaler
SC = StandardScaler()
X_train = SC.fit_transform(X_train)
X_test = SC.transform(X_test)
```

### Train Set

### Model Creation

```python
!pip install lightgbm
```

```python
## import xgboost regressor algorithm and fit the model
from lightgbm import LGBMRegressor

# Default parameters ( n_estimators=100,      -     Number of trees
                    # learning_rate=0.1,    -     Step size for updating
                    # num_leaves=31,        -     Maximum number of leaves in one tree
                    # max_depth=-1,         -    Maximum depth of the tree (-1 means no limit)
                    # min_data_in_leaf=20,  -    Minimum number of data points in a leaf
                    # subsample=0.8          -    Fraction of samples used for fitting )
LGradient = LGBMRegressor()
LGradient.fit(X_train, Y_train)
```

### Test Set

### Evaluation Metrics

```python
Y_pred = LGradient.predict(X_test)
```

```python
from sklearn import metrics
print('R2_Score:',metrics.r2_score(Y_test,Y_pred))
```

### Save the Model

```python
import pickle 
filename = "Final_LGradient_Model.sav"
```

```python
pickle.dump(LGradient,open(filename,'wb'))
```

### Load the Model

```python
load_model = pickle.load(open("Final_LGradient_Model.sav",'rb'))
Result = load_model.predict([[21,30.5,2,0,1]])
```
