# Machine Learning
## Decision Tree
### Supports both Regression & Classification Algorithm 

```python
import pandas as pd
```

### Data Collection

```python
dataset = pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

### Input & Output Split

```python
independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
dependent = dataset[['Profit']]
```

```python
independent
```

```python
dependent
```

### Train & Test set split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.20,random_state=0)
```

## Train Set

### Model Creation - Phase 1

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion='squared_error', splitter='best', max_features='sqrt')
regressor.fit(x_train,y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

## Test Set

### Evaluvation Metrics

```python
y_pred = regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
R2_score = r2_score(y_test,y_pred)
R2_score
```

### Save the best model

```python
import pickle
filename = "Final_DecisionTree_model.sav"
```

```python
pickle.dump(regressor,open(filename,'wb'))
```

### Load the best model and check for the prediction

```python
load_model = pickle.load(open("Final_DecisionTree_model.sav",'rb'))
result=load_model.predict([[12365,23654,25896,0,1]])
result
```
